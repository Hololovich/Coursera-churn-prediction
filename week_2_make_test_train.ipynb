```python
'''1. OneHotEncoding - замена одного категориального признака бинарными. Количество бинарных признаков равно количеству уникальных категорий (+1, если присутствуют пропуски)
2. LabelEncoder - замена категорий числами, соответствующими номеру уникального значения категории. (способ плохой, так как теряется смысл категориальности: значения могут неправильно интерпретироваться)
3. BinaryEncoder - категории кодируются как порядковые, числа переводятся в бинарный вид, затем цифры из этой двоичной строки разбиваются на отдельные столбцы. Чем-то похоже на One-Hot, но с меньшей размерностью и искажением расстояния.
4. Замена каждой категории числом входящих в нее объектов'''
```

```python
'''С помощью F-меры с параметром бета. Данная метрика позволит учесть несбалансированность выборки и цену ошибок разного рода.
Параметр бета равным 0.5, т.е. для алгоритма будет важнее большее значение полноты(полнота для нашей задачи означает насколько точно наш алгоритм смог определить класс "отток". Определить класс "отток" будет важнее, чем неправильно идентифицировать класс "не отток"). '''
```

```python
'''будем рассматривать AUC-PRC (площадь под PRC кривой). Это позволит лучше интерпретировать основную метрику, хорошо подходит для несбалансированной выборки.'''
```
